### Outliers

```python
rawData[rawData["minimum_nights"] > 180]
```

```python
rawData[rawData["reviews_per_month"] > 30]
```

```python
filteredData = rawData[rawData.reviews_per_month < 35]
```

```python
filteredData = filteredData[filteredData.minimum_nights < 180]
filteredData = filteredData[filteredData.price < 1000]
```

```python
filteredData.describe()
```

```python
filteredData.hist(bins=20, figsize=(20,15))
plt.show()
```

```python
filteredData.plot(kind="scatter", x="latitude", y="longitude", s=0.05, alpha=0.1)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.DataFrame(filteredData, columns=['price'])
```

```python
#print(df)
df.head()
```

```python
kmeans = KMeans(n_clusters = 10)
```

```python
theClusters = kmeans.fit_predict(df[['price']])
```

```python
print(theClusters)
```

```python
df['ClusterNumber'] = theClusters
```

```python
print(df)
```

```python
df['y'] = 0
```

```python
print(df)
```

```python
plt.scatter(df["price"], df["y"], c=df.ClusterNumber.astype(float), s=1)
plt.show()
```
